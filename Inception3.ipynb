```
import keras
```

```
from keras.applications import inception_v3
```

```
model=inception_v3.InceptionV3()
```

```
model.input
```

```
model.summary()
```

```
from keras.preprocessing import image
```

```
im=image.load_img(path="labrador.jpg",target_size=(299,299))
```

```
im
```

```
type(im)
```

```
import numpy as np
```

```
kim=np.array(im)
```

```
kim.shape
```

```
fim=np.expand_dims(kim,axis=0)
```

```
fim.shape
```

```
from keras.applications.inception_v3 import decode_predictions
```

```
from keras.applications.inception_v3 import preprocess_input
```

```
pfim=preprocess_input(fim)
```

```
pfim.shape
```

```
decode_predictions(model.predict(pfim),top=3)
```
